# **ประเมินคุณภาพโมเดลทั้งสอง**

```python
alogo = [
    [RandomForestClassifier(), 'RandomForestClassifier'],
    [GradientBoostingClassifier(), 'GradientBoostingClassifier'],
    [BaggingClassifier(),'BaggingClassifier'],
    [AdaBoostClassifier(),'AdaBoostClassifier'],
]
model_score = []
for a in alogo :
    model = a[0]
    model.fit(X_train,y_train)
    score = model.score(X_test, y_test)
    model_score.append([score, a[1]])
    y_predicted = model.predict(X_test)
    print(f'{a[1]:20} score : {score:.04f}')
    print(metrics.confusion_matrix(y_test,y_predicted))
    print(metrics.classification_report(y_test, y_predicted))
    print('-'*100)

print(model_score)
print(f'best score = {max(model_score)}')
```

# **Hyperparameter_tuning RF**




```python
#Hyperparameter tuning model_RF
random_state_range = np.arange(1,12,1)
n_estimator_range = np.arange(10,210,10)
param_grid = dict(random_state=random_state_range, n_estimators =n_estimator_range)

rf = RandomForestClassifier()


grid_rf = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5 )
```

```python
np.arange(1,11,1)
```

```python
grid_rf.fit(X_train, y_train)
```

```python
print("the best hyperparameters RF model are %s with a score of %0.2f"
      % (grid_rf.best_params_, grid_rf.best_score_))
```

# **dataframe of grid search parameter rf**

```python
grid_rf_results = pd.concat([pd.DataFrame(grid_rf.cv_results_["params"]),
                          pd.DataFrame(grid_rf.cv_results_["mean_test_score"],
                                       columns=["Accuracy"])],axis=1)
grid_rf_results.head()
```

```python
grid_rf_contour = grid_rf_results.groupby(['random_state','n_estimators']).mean()
grid_rf_contour
```

# **pivoting the data RF**

```python
grid_rf_reset = grid_rf_contour.reset_index()
grid_rf_reset.columns = ['random_state', 'n_estimators', 'Accuracy']
grid_rf_pivot = grid_rf_reset.pivot('random_state', 'n_estimators')
grid_rf_pivot
```

```python
x_rf = grid_rf_pivot.columns.levels[1].values
y_rf = grid_rf_pivot.index.values
z_rf = grid_rf_pivot.values
```

# **2D Contour Plot RF**

```python
import plotly.graph_objects as go

# X and Y axes labels
layout = go.Layout(
            xaxis=go.layout.XAxis(
              title=go.layout.xaxis.Title(
              text='n_estimators')
             ),
             yaxis=go.layout.YAxis(
              title=go.layout.yaxis.Title(
              text='random_state')
            ) )

fig = go.Figure(data = [go.Contour(z=z_rf, x=x_rf, y=y_rf)], layout=layout )

fig.update_layout(title='Hyperparameter tuning', autosize=False,
                  width=500, height=500,
                  margin=dict(l=65, r=50, b=65, t=90))

fig.show()
```

# **3D Surface Plot RF**

```python
import plotly.graph_objects as go


fig = go.Figure(data= [go.Surface(z=z_rf, y=y_rf, x=x_rf)], layout=layout )
fig.update_layout(title='Hyperparameter tuning',
                  scene = dict(
                    xaxis_title='n_estimators',
                    yaxis_title='random_state',
                    zaxis_title='Accuracy'),
                  autosize=False,
                  width=800, height=800,
                  margin=dict(l=65, r=50, b=65, t=90))
fig.show()
```
