```python
import matplotlib.pyplot as plt
from astropy.coordinates import SkyCoord
import astropy.units as u
import matplotlib.image as mpimg

image_path = 'static/img/plot.jpg'
image = mpimg.imread(image_path)

# fig, ax = plt.subplots(figsize=(8, 6))


# Add this magic command to display the plot in Jupyter Notebook or JupyterLab
# %matplotlib inline

# Define sky coordinates
ra = [50.75] * u.deg
dec = [43.22] * u.deg

sky_coords = SkyCoord(dec, ra)

# Create a Matplotlib figure and axis
fig, ax = plt.subplots(figsize=(8, 6))
ax.imshow(image)

# Plot the sky coordinates
ax.scatter(sky_coords.ra.deg, sky_coords.dec.deg, marker='o', color='blue')

# Add labels, legends, and titles
ax.set_xlabel('Right Ascension (deg)')
ax.set_ylabel('Declination (deg)')
ax.set_title('Sky Coordinate-Aware Plotting Example')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Load the image
image_path = 'static/img/plot.jpg'
image = mpimg.imread(image_path)

# Create a Matplotlib figure and axis
fig, ax = plt.subplots(figsize=(8, 6))

# Set the background color of the figure
fig.patch.set_facecolor('black')

# Set the background color of the axis
ax.set_facecolor('black')

# Display the image as a background using imshow()
ax.imshow(image)

# Plot additional data on top of the image
x = [1.5]  # Sample x-coordinate
y = [2.7]  # Sample y-coordinate
ax.scatter(x, y, color='red', marker='o', label='Prediction Position')

# Add labels, legends, and titles

ax.set_xlabel('Right Ascension (deg)')
ax.set_ylabel('Declination (deg)')
title = ax.set_title('Sky Coordinate-Aware Plotting Example')
ax.set_title('Sky Coordinate-Aware Plotting Example' )
ax.set_title.set_color('blue')

title = ax.set_title('Sky Coordinate')
title.set_color('white')
title

xlabel = ax.set_xlabel('X-axis')
xlabel.set_color('green')

# Set the y-axis label and change the color
ylabel = ax.set_ylabel('Y-axis')
ylabel.set_color('red')

# Show the legend
ax.legend()

# Remove the axis ticks and labels if desired
ax.set_xticks([])
ax.set_yticks([])

# Display the plot
plt.show()
```
