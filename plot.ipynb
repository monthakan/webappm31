```python
import matplotlib.pyplot as plt
from astropy.coordinates import SkyCoord
import astropy.units as u
import matplotlib.image as mpimg

image_path = 'static/img/plot.jpg'
image = mpimg.imread(image_path)

# fig, ax = plt.subplots(figsize=(8, 6))


# Add this magic command to display the plot in Jupyter Notebook or JupyterLab
# %matplotlib inline

# Define sky coordinates
ra = [50.75] * u.deg
dec = [43.22] * u.deg

sky_coords = SkyCoord(dec, ra)

# Create a Matplotlib figure and axis
fig, ax = plt.subplots(figsize=(8, 6))
height, width, _ = image.shape
ax.imshow(image, origin='upper', extent=[0, width, 0, height]) 


# Plot the sky coordinates
ax.scatter(sky_coords.ra.deg, sky_coords.dec.deg, marker='o', color='blue')

# Add labels, legends, and titles
ax.set_xlabel('Right Ascension (deg)')
ax.set_ylabel('Declination (deg)')
ax.set_title('Sky Coordinate-Aware Plotting Example')
ax.set_ylim([0,1800])

# Display the plot
plt.show()
```
